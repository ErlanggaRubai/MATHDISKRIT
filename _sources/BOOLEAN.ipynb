## Operator Biner
Operator biner adalah operator yang mengoperasikan dua operan dan memanipulasinya untuk menghasilkan hasil. Operator direpresentasikan oleh karakter khusus atau kata kunci dan menyediakan cara mudah untuk membandingkan nilai numerik atau rangkaian karakter.

## Boolean Algebra
Boolean Algebra adalah cabang aljabar yang berhubungan dengan operasi logika pada variabel yang hanya memiliki dua nilai, yaitu benar (True) dan salah (False), atau biasa ditandai dengan angka 1 dan 0. Ditemukan oleh matematikawan George Boole, Boolean Algebra menjadi dasar dari sistem logika yang digunakan dalam bidang komputasi, elektronik digital, dan pemrograman.
